```python
import csv
import random

csvfile = open('./Domestic Exchange.csv')
reader = csv.DictReader(csvfile)
GG = dict()
print('{')
print('"nodes": [')
for row in reader:
    colorgen = "%06x" % random.randint(0, 0xFFFFFF)
    if row['ASN-source'] not in GG:
        GG[row['ASN-source']] = 0
    else:
        GG[row['ASN-source']] += 1
    print('{')
    print('"color": "{}",'.format(colorgen))
    print('"label": "{}",'.format(row['ASN-source']))
    print('"attributes": {},')
    print('"x": {},'.format(random.randint(-1000,1000)))
    print('"y": {},'.format(random.randint(-1000,1000)))
    print('"id": "{}",'.format(row['ASN']))
    print('"size": {}'.format(GG[row['ASN-source']]))
    print('},')
print('],')
print('"edges": [')
for row in reader:  
    print('{')
    print('"sourceID": "{}",'.format(row['ASN-source']))
    print('"attributes": {},')
    print('"targetID": "{}",'.format(row['ASN']))
    print('"size": {}'.format(row['Bandwidth']/10))
    print('},')
print(']')
print('}')
    
```
